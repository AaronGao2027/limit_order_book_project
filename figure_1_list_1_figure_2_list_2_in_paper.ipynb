Listing 1 & Table 1

```python
query = f"""
SELECT * 
FROM 'spy_mbo/processed/spy_mbo.parquet'
WHERE ts_event >= '2026-03-23 10:00:00' AND ts_event <= '2026-03-23 10:30:00'
"""
market_by_order_df = duckdb.sql(query).df()
market_by_order_df
```
